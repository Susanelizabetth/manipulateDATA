# TABLE RESUMEN

CHANGES AND EDIT RESUMEN TABLE

```python
df_new = df.rename(columns={'RESUMEN' : 'EVENTO',
                            'Unnamed: 1' : 'FOTOS',
                            'Unnamed: 2' : 'POST',
                            'Unnamed: 3' : 'VIDEOS',
                            'Unnamed: 4' : 'MENSAJE',
                            'Unnamed: 5' : 'NOTICIAS',
                            'Unnamed: 6' : 'CHECK'})
#drop total
df_new.drop([0,306], inplace = True)

#drop NaN
nan_value = float("NaN")

df_new.replace("", nan_value, inplace=True)

df_new.dropna(subset = ['EVENTO'], inplace = True)

df_new.replace(np.nan, 0, inplace = True)

df_new['CHECK'] = df_new['CHECK'].replace(0, 'LISTO')

df_new = df_new.reset_index()
df_new.drop(columns=['index'], inplace=True)

df_new
```

# UTILS FUNCTIONS

```python
def CleanerFunction(data, name):
    #drop 0 row
    data.drop([0], inplace = True)

    #drop NaN
    nan_value = float("NaN")
    data.replace("", nan_value, inplace=True)
    data.dropna(subset = [name], inplace = True)

    data.drop(index=data.index[-1],axis=0,inplace=True)
    
    #replace NaN
    data['CHECK'] = data['CHECK'].replace(np.nan, 'N')
    return(data)
```

```python
#cleaner df2
df_foto = df2[['Unnamed: 0', 'Unnamed: 1', 'Unnamed: 2']].copy()

label = 'FOTOGRAFÍA'

df_foto.rename(columns={'Unnamed: 0' : label,
                        'Unnamed: 1' : 'CANTIDAD',
                        'Unnamed: 2' : 'CHECK'}, inplace = True)


df_foto = CleanerFunction(df_foto, label)

df_foto

```
