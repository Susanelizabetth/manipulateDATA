# UTILS FUNCTIONS

```python
def CleanerFunction(data, name):
    #drop 0 row
    data.drop([0], inplace = True)

    #drop NaN
    nan_value = float("NaN")
    data.replace("", nan_value, inplace=True)
    data.dropna(subset = [name], inplace = True)

    data.drop(index=data.index[-1],axis=0,inplace=True)
    
    #replace NaN
    data['CHECK'] = data['CHECK'].replace(np.nan, 'N')
    
    data = data.reset_index()
    data.drop(columns=['index'], inplace=True)
    
    return(data)
```

```python
def check(data, i):
    data['CHECK'][i] = 'S'
```

```python
def SearchFunc(text, data, column ):
    eventsRes = data[column].tolist()
    name = [text]*len(eventsRes)
    
    if column == 'EVENTO':
        list_search = list(zip(eventsRes,name))
        df_search = pd.DataFrame(list_search)
        df_search.rename(columns={0 : 'RESUMEN',
                                  1 : 'SEARCH'}, inplace = True)
    else:
        quantity = data['CANTIDAD'].tolist()
        check = data['CHECK'].tolist()
        list_search = list(zip(eventsRes,name,quantity,check))
        df_search = pd.DataFrame(list_search)
        df_search.rename(columns={0 : 'RESUMEN',
                                  1 : 'SEARCH',
                                  2 : 'CANTIDAD',
                                  3 : 'CHECK'}, inplace = True)
        
    r = []

    for row in df_search.index:
        x = df_search['RESUMEN'][row]
        y = df_search['SEARCH'][row]
        ratio = SequenceMatcher(None, y, x).ratio()
        r.append(ratio)

    df_search.insert(2, "RATIO", r, True)    

    df_search = df_search.loc[df_search['RATIO']> 0.5]
    return(df_search)
```
